```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read both CSV files
df_uc_flux = pd.read_csv("../../data/UC-UN/UC-diameter-flux.csv", skiprows=1)
df_uc_un_uo2 = pd.read_csv("../../data/UC-UN/UC-UN-UO2-diameter.csv", skiprows=1)


# Diameter-UC-300: columns 4,5
uc_300_file1_x_raw = pd.to_numeric(df_uc_flux.iloc[:, 4], errors='coerce').dropna()
uc_300_file1_y_raw = pd.to_numeric(df_uc_flux.iloc[:, 5], errors='coerce').dropna()
min_len_300_f1 = min(len(uc_300_file1_x_raw), len(uc_300_file1_y_raw))
# Sort by X values and apply same sorting to Y
sort_idx_300_f1 = np.argsort(uc_300_file1_x_raw.iloc[:min_len_300_f1].values)
uc_300_file1_x = uc_300_file1_x_raw.iloc[:min_len_300_f1].values[sort_idx_300_f1]
uc_300_file1_y = uc_300_file1_y_raw.iloc[:min_len_300_f1].values[sort_idx_300_f1]

# Diameter-UC-600: columns 6,7
uc_600_file1_x_raw = pd.to_numeric(df_uc_flux.iloc[:, 6], errors='coerce').dropna()
uc_600_file1_y_raw = pd.to_numeric(df_uc_flux.iloc[:, 7], errors='coerce').dropna()
min_len_600_f1 = min(len(uc_600_file1_x_raw), len(uc_600_file1_y_raw))
# Sort by X values and apply same sorting to Y
sort_idx_600_f1 = np.argsort(uc_600_file1_x_raw.iloc[:min_len_600_f1].values)
uc_600_file1_x = uc_600_file1_x_raw.iloc[:min_len_600_f1].values[sort_idx_600_f1]
uc_600_file1_y = uc_600_file1_y_raw.iloc[:min_len_600_f1].values[sort_idx_600_f1]

# Extract and sort UC data from second file (UC-UN-UO2-diameter.csv)
# UC-diameter-600: columns 0,1
uc_600_file2_x_raw = pd.to_numeric(df_uc_un_uo2.iloc[:, 2], errors='coerce').dropna()
uc_600_file2_y_raw = pd.to_numeric(df_uc_un_uo2.iloc[:, 3], errors='coerce').dropna()
min_len_600_f2 = min(len(uc_600_file2_x_raw), len(uc_600_file2_y_raw))
# Sort by X values and apply same sorting to Y
sort_idx_600_f2 = np.argsort(uc_600_file2_x_raw.iloc[:min_len_600_f2].values)
uc_600_file2_x = uc_600_file2_x_raw.iloc[:min_len_600_f2].values[sort_idx_600_f2]
uc_600_file2_y = uc_600_file2_y_raw.iloc[:min_len_600_f2].values[sort_idx_600_f2]

# UC-diameter-300: columns 2,3
uc_300_file2_x_raw = pd.to_numeric(df_uc_un_uo2.iloc[:, 0], errors='coerce').dropna()
uc_300_file2_y_raw = pd.to_numeric(df_uc_un_uo2.iloc[:, 1], errors='coerce').dropna()
min_len_300_f2 = min(len(uc_300_file2_x_raw), len(uc_300_file2_y_raw))
# Sort by X values and apply same sorting to Y
sort_idx_300_f2 = np.argsort(uc_300_file2_x_raw.iloc[:min_len_300_f2].values)
uc_300_file2_x = uc_300_file2_x_raw.iloc[:min_len_300_f2].values[sort_idx_300_f2]
uc_300_file2_y = uc_300_file2_y_raw.iloc[:min_len_300_f2].values[sort_idx_300_f2]


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))

# UC-300 comparison
ax1.plot(uc_300_file1_x, uc_300_file1_y, 'o-', color='blue', label='Fig 1', markersize=8, alpha=0.7)
ax1.plot(uc_300_file2_x, uc_300_file2_y, 's-', color='red', label='Fig 2', markersize=8, alpha=0.7)
ax1.set_title('UC-300°C Diameter Comparison (Sorted)', fontsize=12, fontweight='bold')
ax1.set_xlabel('X Value')
ax1.set_ylabel('Diameter (nm)')
ax1.legend()
ax1.grid(True, alpha=0.3)

# UC-600 comparison
ax2.plot(uc_600_file1_x, uc_600_file1_y, 'o-', color='blue', label='Fig 1', markersize=8, alpha=0.7)
ax2.plot(uc_600_file2_x, uc_600_file2_y, 's-', color='red', label='Fig 2', markersize=8, alpha=0.7)
ax2.set_title('UC-600°C Diameter Comparison (Sorted)', fontsize=12, fontweight='bold')
ax2.set_xlabel('X Value')
ax2.set_ylabel('Diameter (nm)')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Overlay plots for better comparison
ax3.plot(uc_300_file1_x, uc_300_file1_y, 'o-', color='blue', label='UC-300 fig 1', markersize=6, alpha=0.7)
ax3.plot(uc_300_file2_x, uc_300_file2_y, 's-', color='red', label='UC-300 fig 2', markersize=6, alpha=0.7)
ax3.plot(uc_600_file1_x, uc_600_file1_y, '^-', color='green', label='UC-600 fig 1', markersize=6, alpha=0.7)
ax3.plot(uc_600_file2_x, uc_600_file2_y, 'D-', color='orange', label='UC-600 fig 2', markersize=6, alpha=0.7)
ax3.set_title('All UC Data Overlay (Sorted)', fontsize=12, fontweight='bold')
ax3.set_xlabel('X Value')
ax3.set_ylabel('Diameter (nm)')
ax3.legend()
ax3.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
