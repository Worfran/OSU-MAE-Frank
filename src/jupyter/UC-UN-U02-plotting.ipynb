```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the CSV file
df = pd.read_csv("../../data/UC-UN/UC-UN-UO2-diameter.csv", skiprows=1)

# Extract and sort data for each material and temperature
# UC data (columns 0-3)
# UC-diameter-600: columns 0,1
uc_600_x_raw = pd.to_numeric(df.iloc[:, 0], errors='coerce').dropna()
uc_600_y_raw = pd.to_numeric(df.iloc[:, 1], errors='coerce').dropna()
min_len_uc_600 = min(len(uc_600_x_raw), len(uc_600_y_raw))
sort_idx_uc_600 = np.argsort(uc_600_x_raw.iloc[:min_len_uc_600].values)
uc_600_x = uc_600_x_raw.iloc[:min_len_uc_600].values[sort_idx_uc_600]
uc_600_y = uc_600_y_raw.iloc[:min_len_uc_600].values[sort_idx_uc_600]

# UC-diameter-300: columns 2,3
uc_300_x_raw = pd.to_numeric(df.iloc[:, 2], errors='coerce').dropna()
uc_300_y_raw = pd.to_numeric(df.iloc[:, 3], errors='coerce').dropna()
min_len_uc_300 = min(len(uc_300_x_raw), len(uc_300_y_raw))
sort_idx_uc_300 = np.argsort(uc_300_x_raw.iloc[:min_len_uc_300].values)
uc_300_x = uc_300_x_raw.iloc[:min_len_uc_300].values[sort_idx_uc_300]
uc_300_y = uc_300_y_raw.iloc[:min_len_uc_300].values[sort_idx_uc_300]

# UN data (columns 4-7)
# UN-diameter-700: columns 4,5
un_700_x_raw = pd.to_numeric(df.iloc[:, 4], errors='coerce').dropna()
un_700_y_raw = pd.to_numeric(df.iloc[:, 5], errors='coerce').dropna()
min_len_un_700 = min(len(un_700_x_raw), len(un_700_y_raw))
sort_idx_un_700 = np.argsort(un_700_x_raw.iloc[:min_len_un_700].values)
un_700_x = un_700_x_raw.iloc[:min_len_un_700].values[sort_idx_un_700]
un_700_y = un_700_y_raw.iloc[:min_len_un_700].values[sort_idx_un_700]

# UN-diameter-800: columns 6,7
un_800_x_raw = pd.to_numeric(df.iloc[:, 6], errors='coerce').dropna()
un_800_y_raw = pd.to_numeric(df.iloc[:, 7], errors='coerce').dropna()
min_len_un_800 = min(len(un_800_x_raw), len(un_800_y_raw))
sort_idx_un_800 = np.argsort(un_800_x_raw.iloc[:min_len_un_800].values)
un_800_x = un_800_x_raw.iloc[:min_len_un_800].values[sort_idx_un_800]
un_800_y = un_800_y_raw.iloc[:min_len_un_800].values[sort_idx_un_800]

# UO2 data (columns 8-11)
# UO2-diameter-600: columns 8,9
uo2_600_x_raw = pd.to_numeric(df.iloc[:, 8], errors='coerce').dropna()
uo2_600_y_raw = pd.to_numeric(df.iloc[:, 9], errors='coerce').dropna()
min_len_uo2_600 = min(len(uo2_600_x_raw), len(uo2_600_y_raw))
sort_idx_uo2_600 = np.argsort(uo2_600_x_raw.iloc[:min_len_uo2_600].values)
uo2_600_x = uo2_600_x_raw.iloc[:min_len_uo2_600].values[sort_idx_uo2_600]
uo2_600_y = uo2_600_y_raw.iloc[:min_len_uo2_600].values[sort_idx_uo2_600]

# UO2-diameter-800: columns 10,11
uo2_800_x_raw = pd.to_numeric(df.iloc[:, 10], errors='coerce').dropna()
uo2_800_y_raw = pd.to_numeric(df.iloc[:, 11], errors='coerce').dropna()
min_len_uo2_800 = min(len(uo2_800_x_raw), len(uo2_800_y_raw))
sort_idx_uo2_800 = np.argsort(uo2_800_x_raw.iloc[:min_len_uo2_800].values)
uo2_800_x = uo2_800_x_raw.iloc[:min_len_uo2_800].values[sort_idx_uo2_800]
uo2_800_y = uo2_800_y_raw.iloc[:min_len_uo2_800].values[sort_idx_uo2_800]
```

```python
# Create comparison plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# UC vs UN comparison
ax1.plot(uc_300_x, uc_300_y, 'o-', color='#1f77b4', label='UC-300°C', 
         markersize=7, alpha=0.8, linewidth=2)
ax1.plot(uc_600_x, uc_600_y, 's-', color='#ff7f0e', label='UC-600°C', 
         markersize=7, alpha=0.8, linewidth=2)
ax1.plot(un_700_x, un_700_y, '^-', color='#2ca02c', label='UN-700°C', 
         markersize=7, alpha=0.8, linewidth=2)
ax1.plot(un_800_x, un_800_y, 'D-', color='#d62728', label='UN-800°C', 
         markersize=7, alpha=0.8, linewidth=2)

ax1.set_title('UC vs UN: Diameter Comparison', fontsize=14, fontweight='bold')
ax1.set_xlabel('Dose (dpa)', fontsize=12)
ax1.set_ylabel('Diameter (nm)', fontsize=12)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# UC vs UO2 comparison
ax2.plot(uc_300_x, uc_300_y, 'o-', color='#1f77b4', label='UC-300°C', 
         markersize=7, alpha=0.8, linewidth=2)
ax2.plot(uc_600_x, uc_600_y, 's-', color='#ff7f0e', label='UC-600°C', 
         markersize=7, alpha=0.8, linewidth=2)
ax2.plot(uo2_600_x, uo2_600_y, '^-', color='#9467bd', label='UO₂-600°C', 
         markersize=7, alpha=0.8, linewidth=2)
ax2.plot(uo2_800_x, uo2_800_y, 'D-', color='#8c564b', label='UO₂-800°C', 
         markersize=7, alpha=0.8, linewidth=2)

ax2.set_title('UC vs UO₂: Diameter Comparison', fontsize=14, fontweight='bold')
ax2.set_xlabel('Dose (dpa)', fontsize=12)
ax2.set_ylabel('Diameter (nm)', fontsize=12)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
